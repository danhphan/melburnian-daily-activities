```python
import streamlit as st
import pandas as pd
import geopandas as gpd
import numpy as np
import pydeck as pdk
```

```python
print("Welcome to pydeck version", pdk.__version__)
```

```python
ACTS_DATA = "../data/activities.geojson"
acts = gpd.read_file(ACTS_DATA)
acts.shape
```

```python
acts = acts.to_crs(epsg=4326)
```

```python
acts.head()
```

```python
view = pdk.ViewState(latitude=-37.7989, longitude=144.9591, pitch=50, zoom=7)
layers = [
        pdk.Layer(
            'HexagonLayer',
            acts,
            get_position="geometry.coordinates",
            auto_highlight=True,
            elevation_scale=50,
            pickable=True,
            elevation_range=[0, 3000],
            extruded=True,                 
            coverage=1)
        ]

r = pdk.Deck(layers, initial_view_state=view)
```
